# 加速神经网络训练 (Speed Up Training)

包括以下几种模式：
- Stochastic Gradient Descent (SGD)
- Momentum
- AdaGrad
- RMSProp
- Adam

越复杂的神经网络 , 越多的数据 , 我们需要在训练神经网络的过程上花费的时间也就越多. 原因很简单, 就是因为计算量太大了. 可是往往有时候为了解决复杂的问题, 复杂的结构和大数据又是不能避免的, 所以我们需要寻找一些方法, 让神经网络聪明起来, 快起来.

# Stochastic Gradient Descent (SGD)

最基础地方法时SGD，假设红色方块就是我们要训练的data，如果使用普通的训练方法，我们就需要不断地将整套数据放入神经网络NN中进行训练，这样消耗地资源会比较大。

如果我们换个思路，我们将这些数据拆分成一小批一小批的，然后再分批不断地放入神经网络NN中进行计算，这就是我们常说的SDG的正确打开方式了。每次使用批数据虽然不能反映整体数据的情况，不过却很大程度上加速了NN的训练过程，而且也不会丢失太多准确率。

# Momentum 更新方法 

大多数其他途径是再更新神经网络参数的时候动动手脚。传统的参数W的更新是把原始的W累加上一个负的学习率（learning rate）乘以校正值（dx）。这种方法可能会让学习过程曲折无比，看起来像一个喝醉的人回家时，摇摇晃晃走了很多弯路。

所以我们把这个人从平地上放到一个斜坡上，只要他往下坡的方向走一点点，由于向下的惯性，他不自觉地就一直往下走，走地弯路也变少了。这就是Momentum参数更新。另外一种加速方法叫AdaGrad。

# AdaGrad更新方法

这种方法是在学习率上面动手脚，使得每一个参数更新都会有自己与众不同地学习率，他的作用和momentum类似，不过不是给喝酒的人安排另一个下坡，而是给他一双不好走路的鞋子，使得他一摇晃着走路就脚疼，鞋子成为了走弯路的阻力，逼着他往前直着走。他的数学形式是这样的。

入股我们将上述两种方法结合起来是不是更好呢？这样就有了RMSProp更新方法。

# RMSProp更新方法

有了momentum的惯性原则，加上adagrad的对错误方向的阻力，我们就能合并成这样，让RMSProp同时具备他们两种方法的优势。

不过RMSProp方法还没把Momentum合并完全，RMSProp还缺少了momentum中的这一部分。所以，我们在Adam方法中补上了这种想法。

# Adam更新方法

计算m时有momentum下坡的属性，计算v时有adagrad阻力的属性，然后再更新参数时把m和v都考虑进去。实验证明，大多数时候，使用adam都能又快又好地达到目标，迅速收敛。
